```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('Directory.csv')
df
```

```python
X = df[['JobRole', 'WorkLocation']]
y = df['Salary']
# X_numeric = pd.get_dummies(X)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# model = LinearRegression()
# model.fit(X_numeric,y)
# y_pred = model.predict(X_test)
```

```python
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

preprocessor = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), ['JobRole', 'WorkLocation'])])
pipeline = Pipeline(steps=[('preprocessor', preprocessor), ('regressor', LinearRegression())])

pipeline.fit(X_train, y_train)
y_pred = pipeline.predict(X_test)

distinct_combinations = np.unique(y_pred)
print("Predicted salary: ")
# for i in range(len(y_test)):
#     print(f"Job Role:{X_test.iloc[i]['JobRole']}, Work Location: {X_test.iloc[i]['WorkLocation']}, Predicted Salary: {y_pred[i]:.2f}")
# for salary in distinct_combinations:
#     print(f"Job Role:{X_test.iloc[i]['JobRole']}, Work Location: {X_test.iloc[i]['WorkLocation']}, Predicted Salary: {salary:.2f}")
```

```python
plt.figure(figsize=(30, 20))
plt.plot(range(len(y_test)), y_test, marker = 'o', label = 'Actual Salary')
plt.plot(range(len(y_test)), y_pred, marker = 'o', label = 'Predicted Salary', color='tab:orange', alpha=0.5)
plt.xlabel('Data Point')
plt.ylabel('Salary')
plt.title('Actual vs. Predicted Salary')
plt.legend()
plt.show()
```
